```python
import pandas as pd
import ipaddress
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
import numpy as np
import joblib
from sklearn.ensemble import IsolationForest

pd.set_option('future.no_silent_downcasting', True)
column_names = ['ts', 'uid', 'id.orig_h', 'id.orig_p', 'id.resp_h', 'id.resp_p', 'proto', 'service', 'duration', 'orig_bytes', 'resp_bytes', 'conn_state', 'local_orig', 'local_resp', 'missed_bytes', 'history', 'orig_pkts', 'orig_ip_bytes', 'resp_pkts', 'resp_ip_bytes', 'tunnel_parents', 'ip_proto']
zeek_data = pd.read_csv('/opt/homebrew/opt/zeek/logs/data.tsv', delimiter='\t', comment='#', names=column_names)
zeek_data
```

```python
zeek_data = zeek_data.dropna(axis=1, how='all')  # Drop columns with all NaN values
zeek_data['ts'] = pd.to_datetime(zeek_data['ts'], unit='s')
zeek_data
```

```python
# Drop rows with missing values
zeek_data = zeek_data.dropna()

# Or fill missing values with a placeholder, for example, 0
zeek_data = zeek_data.fillna(0)
```

```python
zeek_data.columns = zeek_data.columns.str.strip()  # Remove any leading/trailing spaces
zeek_data.columns = zeek_data.columns.str.lower()  # Convert all column names to lowercase
zeek_data
```

```python
zeek_data.replace(['...', '-'], pd.NA, inplace=True)
zeek_data.fillna(0, inplace=True)
zeek_data
```

```python
zeek_data['duration'] = pd.to_numeric(zeek_data['duration'], errors='coerce')  # Convert duration to numeric (coerce invalid values to NaN)
zeek_data['orig_bytes'] = pd.to_numeric(zeek_data['orig_bytes'], errors='coerce')  # Convert orig_bytes to numeric
zeek_data['resp_bytes'] = pd.to_numeric(zeek_data['resp_bytes'], errors='coerce')  # Convert resp_bytes to numeric
```

```python

def standardize_ip(ip):
    try:
        return str(ipaddress.ip_address(ip))
    except ValueError:
        return ip

zeek_data['id.orig_h'] = zeek_data['id.orig_h'].apply(standardize_ip)
zeek_data['id.resp_h'] = zeek_data['id.resp_h'].apply(standardize_ip)

zeek_data
```

```python
zeek_data.info()
```

```python

zeek_data_encoded = pd.get_dummies(zeek_data, columns=['proto', 'service', 'conn_state', 'local_orig', 'local_resp', 'history', 'tunnel_parents','id.orig_h', 'id.resp_h' ], drop_first=True)

zeek_data_encoded.info()
zeek_data_encoded
```

```python
zeek_data_encoded
```

```python
features = zeek_data_encoded.drop(columns=['uid', 'ts'])

# Scale the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)


X_train, X_test = train_test_split(features_scaled, test_size=0.2, random_state=42)

# Build the Sequential Autoencoder Model
learning_rate = 0.01
optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)

model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=(X_train.shape[1],)),  # Input layer size is the number of features
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(X_train.shape[1], activation='sigmoid')  # Output layer should match input size
])
model.compile(optimizer=optimizer, loss='mean_squared_error')

model.fit(X_train, X_train, epochs=50, batch_size=32, validation_data=(X_test, X_test))

reconstructions = model.predict(X_test)
mse = np.mean(np.power(X_test - reconstructions, 2), axis=1)

threshold = np.percentile(mse, 98)

anomalies = mse > threshold
print(f"Anomalies detected: {np.sum(anomalies)}")

anomalous_indices = np.where(anomalies)[0]
anomalous_traffic = zeek_data_encoded.iloc[anomalous_indices]

print(anomalous_traffic)

```

```python
# Extract columns where the value is True for each row
anomalous_traffic_values = anomalous_traffic.loc[:, anomalous_traffic.any(axis=0)]

# Iterate over each row and extract only the values that are True in the corresponding columns
for idx, row in anomalous_traffic_values.iterrows():
    print(f"Anomalous row {idx} (Values that are True):")
    anomalous_values = row[row == True].index.tolist()  # Get column names where the value is True
    print(anomalous_values)  # This will print the column names where anomalies were detected
    print("Values:", row[anomalous_values].tolist())  # This will print the corresponding values
    print("\n")
```
